```python
import pandas as pd

# Read the original file
EXP3 = pd.read_csv("/hpcstor6/scratch01/h/huuthanhvy.nguyen001/EXP3/finetuning-EXP3numbertwo/EXP-Results/EXP3results.csv")

# Let's first check how many unique images are in each task group
print("Initial count:")
for task in EXP3['task_name'].unique():
    task_data = EXP3[EXP3['task_name'] == task]
    unique_images = task_data['image_path'].nunique()
    print(f"{task}: {unique_images} unique images")

# Get first 10 unique images for each task (removing the last unique image)
reduced_df = pd.concat([
    group[group['image_path'].isin(group['image_path'].unique()[:10])]
    for name, group in EXP3.groupby('task_name')
])

# Verify the results
print("\nAfter reduction:")
for task in reduced_df['task_name'].unique():
    task_data = reduced_df[reduced_df['task_name'] == task]
    print(f"\n{task}:")
    print(f"Total rows: {len(task_data)}")  # Should be 150 (10 images Ã— 15 responses per image)
    print(f"Unique images: {task_data['image_path'].nunique()}")  # Should be 10

# Save to new CSV file
output_path = "/hpcstor6/scratch01/h/huuthanhvy.nguyen001/EXP3/finetuning-EXP3numbertwo/EXP-Results/EXP3results10images.csv"
reduced_df.to_csv(output_path, index=False)
```

```python
import pandas as pd

# Define the file paths
file_paths = [
    #"/hpcstor6/scratch01/h/huuthanhvy.nguyen001/EXP4/finetuning-EXP4numberone/EXP-Results/EXP4results.csv",
    #"/hpcstor6/scratch01/h/huuthanhvy.nguyen001/EXP4/finetuning-EXP4numbertwo/EXP-Results/EXP4results.csv",
    "/hpcstor6/scratch01/h/huuthanhvy.nguyen001/EXP4/finetuning-EXP4numberthree/EXP-Results/EXP4results.csv"
]

# Process each file
for input_path in file_paths:
    # Read the file
    df = pd.read_csv(input_path)
    
    print(f"\nProcessing: {input_path}")
    print("Initial count:")
    for task in df['task_name'].unique():
        task_data = df[df['task_name'] == task]
        unique_images = task_data['image_path'].nunique()
        print(f"{task}: {unique_images} unique images")

    # Get first 10 unique images for each task
    reduced_df = pd.concat([
        group[group['image_path'].isin(group['image_path'].unique()[:10])]
        for name, group in df.groupby('task_name')
    ])

    # Verify the results
    print("\nAfter reduction:")
    for task in reduced_df['task_name'].unique():
        task_data = reduced_df[reduced_df['task_name'] == task]
        print(f"\n{task}:")
        print(f"Total rows: {len(task_data)}")
        print(f"Unique images: {task_data['image_path'].nunique()}")

    # Create output path by modifying input path
    output_path = input_path.replace('results.csv', 'results10images.csv')
    
    # Save to new CSV file
    reduced_df.to_csv(output_path, index=False)
    print(f"\nSaved to: {output_path}")
    print("-" * 80)
```
