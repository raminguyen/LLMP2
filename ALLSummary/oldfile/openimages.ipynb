```python
import os
import random
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

# Path to your image directory
img_dir = '/hpcstor6/scratch01/h/huuthanhvy.nguyen001/EXP2/finetuning-EXP2numberone/images'

# List all image files in the directory
all_images = [os.path.join(img_dir, img) for img in os.listdir(img_dir) if img.endswith(('.jpg', '.png'))]

# Select 20 random images
random_images = random.sample(all_images, min(20, len(all_images)))

for img_path in random_images:
    # Open the image
    img = Image.open(img_path)
    
    # Convert to grayscale
    img_gray = img.convert('L')  # 'L' stands for "luminance" which is grayscale
    
    # Invert colors
    img_inverted = np.invert(np.array(img_gray))
    
    # Create a figure
    fig, ax = plt.subplots(1, 1)
    
    # Set figure's background to white
    fig.patch.set_facecolor('white')
    
    # Display the image
    ax.imshow(img_inverted, cmap="gray")
    ax.axis('off')  # Optional: hide axes
    
    # Show the image
    plt.show()
```

```python
import os
import random
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

# Path to your image directory
img_dir = '/hpcstor6/scratch01/h/huuthanhvy.nguyen001/EXP2/finetuning-EXP2numberone/images'

# List all image files in the directory
all_images = [os.path.join(img_dir, img) for img in os.listdir(img_dir) if img.endswith(('.jpg', '.png'))]

# Select 20 random images
random_images = random.sample(all_images, min(20, len(all_images)))

for img_path in random_images:
    # Open the image
    img = Image.open(img_path)
    
    # Convert to grayscale
    img_gray = img.convert('L')  # 'L' stands for "luminance" which is grayscale
    
    # Invert colors for black background and white elements
    img_black_bg = 255 - np.array(img_gray)  # Subtract pixel values from 255 to invert
    
    # Create a figure
    fig, ax = plt.subplots(1, 1)
    
    # Set figure's background to black
    fig.patch.set_facecolor('black')
    
    # Display the image
    ax.imshow(img_black_bg, cmap="gray")
    ax.axis('off')  # Optional: hide axes
    
    # Show the image
    plt.show()
```
