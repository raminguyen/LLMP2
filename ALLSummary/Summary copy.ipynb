```python
from EXP2Summary import checkdeletedrows_forallcsv, process_and_plot_multiplerun, visualize_cleaned_answers_overlay

combined_deleted_df, balanced_dataframes, metrics_table, all_balanced_metrics = checkdeletedrows_forallcsv()

averaged_metrics_table = process_and_plot_multiplerun(metrics_table)

visualize_cleaned_answers_overlay(balanced_dataframes)

averaged_metrics_table
```

```python
from scipy.stats import f_oneway

# Group MLAE values by model
mlae_groups = [group["Average MLAE"].values for _, group in averaged_metrics_table.groupby("Model")]

# Perform one-way ANOVA
f_stat, p_value = f_oneway(*mlae_groups)

# Print results
print("\nOne-Way ANOVA Results for MLAE:")
print(f"F-statistic: {f_stat:.4f}")
print(f"P-value: {p_value:.4e}")

# Interpretation
if p_value < 0.01:
    print("Statistically significant difference in MLAE between models (p < 0.01).")
else:
    print("No significant difference in MLAE between models (p >= 0.01).")
```
